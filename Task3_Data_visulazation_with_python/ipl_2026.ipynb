# create new column 

```python
df['Total_runs'] = df['runs_of_bat']+df['extras']

df['is_six'] = 0
df['is_four'] =0
df['is_wicket']= 0
df.loc[df['wicket_type'].notnull(),'is_wicket']=1

df.loc[df['runs_of_bat'] ==6 , 'is_six' ] = 1 
df.loc[df['runs_of_bat'] ==4 , 'is_four' ] = 1
df.head()
```

```python
df['wicket_type'].unique()
```

```python
df['over'] = df['over'] +1 # change the starting over from 0 to 1
df.head()
```

```python
def get_phase(over):
    if over <= 6:
        return 'powerplay (1-6)'
    elif over <= 15:
        return 'middle (7-15)'
    else:
        return 'death (16-20)'
        
df['phase_group'] = df['over'].apply(get_phase)
df.head()
```

# batting analysis

## most run by team

```python
df.columns
```

```python
# Group by batting team and sum total runs
edf = (
    df.groupby('batting_team', as_index=False)
      .agg(total=('Total_runs', 'sum'))
      .sort_values('total', ascending=False)
      .reset_index(drop=True)
)
edf
```

```python

plt.figure(figsize=(15,6))
plt.bar(edf['batting_team'], edf['total'])
plt.title('Team Wise Total Score')
plt.xlabel('Batting Team')
plt.ylabel('Total Score')

for x, y in zip(edf['batting_team'], edf['total']):
    plt.text(x=x, y=y, s=y, va='bottom',)

plt.show()
```

# team sixes and four

```python
edf = df.groupby('batting_team').agg(
    nos=('is_six', 'sum'),
    nof=('is_four', 'sum')
).reset_index()

edf['score by boundaries'] = edf['nos'] + edf['nof']
edf
```

```python
plt.figure(figsize=(15,6))

xval = np.arange(1,11)

plt.bar(xval, edf['nos'], width=0.3, label='sixes')
plt.bar(xval+0.2, edf['nof'], width=0.3, label='fours')
plt.bar(xval+0.4, edf['score by boundaries'], width=0.3, label='total')

plt.legend()

plt.xticks(xval, edf['batting_team'])
plt.show()
```

#phase wise team run

```python
edf = df.groupby(['batting_team','phase_group']).agg(
    total=('Total_runs','sum')
).reset_index()

edf
```

```python
edf = df.pivot_table(index='batting_team', columns='phase_group',
                     values='Total_runs', aggfunc='sum').reset_index()
edf
```

```python
edf.columns
```

```python
plt.figure(figsize=(20, 6))

xval = np.arange(1, 11)

plt.bar(xval,     edf['powerplay (1-6)'], width=0.3, label='Powerplay')
plt.bar(xval+0.2, edf['middle (7-15)'],   width=0.3, label='Middle')
plt.bar(xval+0.4, edf['death (16-20)'],   width=0.3, label='Death')

plt.legend()
plt.xticks(xval, edf['batting_team'])
plt.show()
```

# top 10 run score

```python
edf = df.groupby('striker').agg(
    total=('Total_runs','sum')
).reset_index()

edf = edf.sort_values('total', ascending=False).head(10)
edf
```

```python
plt.figure(figsize=(20,6))
plt.bar(edf['striker'], edf['total'])
plt.title('Top Strikers')
plt.show()
```

# strike rate

```python
b = 'is_four'
edf = df.groupby('striker').agg(
    total=(b,'sum')
).reset_index()

edf = edf.sort_values('total', ascending=False).head(10)
edf
```

```python
# Exclude wide balls from balls faced
legal_balls_bat = df.loc[df['wide'] == 0]

edf_sr = legal_balls_bat.groupby('striker').agg(
    runs  = ('runs_of_bat', 'sum'),
    balls = ('runs_of_bat', 'count')
).reset_index()

edf_sr = edf_sr.loc[edf_sr['balls'] >= 30]
edf_sr['strike_rate'] = ((edf_sr['runs'] / edf_sr['balls']) * 100).round().astype(int)
edf_sr = edf_sr.sort_values('strike_rate', ascending=False).head(10)
edf_sr
```

```python
plt.figure(figsize=(20,6))
plt.bar(edf['striker'], edf['total'])
plt.show()
```

# top wicket taker 

```python
bowler_wickets = df[(df['wicket_type'].notnull()) & (df['wicket_type']!='runout')]

edf = bowler_wickets.groupby('bowler').agg(
    total=('is_wicket','sum')
).reset_index()

edf = edf.sort_values('total', ascending=False).head(10)
edf
```

```python
plt.figure(figsize=(20,6))
plt.bar(edf['bowler'], edf['total'])
plt.show()
```

# economy rate

```python
plt.figure(figsize=(20,6))
plt.bar(edf['bowler'], edf['economy_rate'])
plt.show()
```

```python
edf = df.groupby('bowler').agg(
    total=('runs_of_bat','sum'),
    balls=('runs_of_bat','count')
).reset_index()

edf = edf[edf['balls'] >= 18]
edf['economy_rate'] = (edf['total']/edf['balls']) * 6

edf = edf.sort_values('economy_rate').head(10)
edf
```

```python
# Exclude wides and no-balls from balls bowled
legal_balls_bowl = df.loc[(df['wide'] == 0) & (df['noballs'] == 0)]

edf_eco = legal_balls_bowl.groupby('bowler').agg(
    total = ('runs_of_bat', 'sum'),
    balls = ('runs_of_bat', 'count')
).reset_index()

edf_eco = edf_eco.loc[edf_eco['balls'] >= 18]
edf_eco['economy_rate'] = ((edf_eco['total'] / edf_eco['balls']) * 6).round(2)
edf_eco = edf_eco.sort_values('economy_rate').head(10)
edf_eco
```
